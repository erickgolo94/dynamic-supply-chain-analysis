```python
# ==============================================================================
# Project: Supply Chain Inventory & Risk Optimizer
# File: Supply_Chain_Inventory_Optimizer_EDA.ipynb
# Author: Erick Andos Golo - Junior Data Analyst
# Created: July 13, 2026
# ==============================================================================
# PROJECT DESCRIPTION & GOAL:
# This project performs an end-to-end Exploratory Data Analysis (EDA) and data 
# cleaning on a comprehensive supply chain dataset (~113k records). 
#
# Primary Objectives:
# 1. Inspect, clean, and handle missing values or anomalies in operational 
#    and risk-related metrics (inventory levels, lead times, shipping costs).
# 2. Aggregate and extract key business insights regarding supplier reliability,
#    route risks, and delay probabilities using standard Pandas operations.
# 3. Export the processed, high-quality dataset for advanced visualization 
#    and interactive dashboarding in Tableau Public.
# 
# Tech Stack: Python, Pandas, NumPy, Tableau
# ==============================================================================

import pandas as pd
import numpy as np

# ==============================================================================
# Phase 1: Load Dataset and Initial Inspection
# ==============================================================================
# Load the raw supply chain dataset into a Pandas DataFrame.
# Ensure the file path points correctly to your local CSV file.
# ==============================================================================

# Load dataset (Update 'supply_chain_data.csv' with your actual file path/name)
#df = pd.read_csv('E:\Projects\Data Analyst\DATA-2026-002_SupplyChain_Inventory_Optimizer\dynamic_supply_chain_logistics_dataset_with_country.csv')
file_path = r'E:\Projects\Data Analyst\DATA-2026-002_SupplyChain_Inventory_Optimizer\01_Raw_Data\dynamic_supply_chain_logistics_dataset_with_country.csv'
df = pd.read_csv(file_path)

#Verify dataset dimensions (Expecting ~113,000+ records)
print("Dataset Shape (Rows, Columns):", df.shape)

# Preview the first 3 rows to inspect raw data values and column formats
display(df.head(3))

## Inspect data types, non-null counts, and memory usage to identify missing values
df.info()
```

```python
# ==============================================================================
# Phase 2: Sanity Checks and Duplicate Inspection
# ==============================================================================
# Check for exact duplicate rows in the dataset
duplicate_count = df.duplicated().sum()
print(f"Number of duplicate rows:{duplicate_count}")

#Generating statistical summary to detect potential logical anomalies (e.g., negative values)
display(df.describe())
```

```python
# ==============================================================================
# Phase 3: Categorical Columns Inspection
# ==============================================================================
# Check unique value counts for key categorical fields
print("Risk Classifications Breakdown:")
print(df['risk_classification'].value_counts())

print("\nNumber of unique supplier countries:", df['supplier_country'].nunique())
```

```python
# ==============================================================================
# Phase 4: Data Aggregation and Export
# ==============================================================================
# Aggregate average shipping costs and delay probability by risk classification

risk_summary = df.groupby('risk_classification').agg({
    'shipping_costs' : 'mean',
    'delay_probability' : 'mean',
    'lead_time_days' : 'mean',
    'product_id' : 'count'
}).rename(columns={'product_id': 'total_shipments'})

print("Summary Metrics by Risk classification:")
display(risk_summary)

#Export the clean dataset to a new CSV file for tableau connection
output_path = r'E:\Projects\Data Analyst\DATA-2026-002_SupplyChain_Inventory_Optimizer\01_Raw_Data\dynamic_supply_chain_logistics_dataset_with_country.csv'
df.to_csv(output_path, index=False)
print(f"\nCleaned dataset successfully exported to: {output_path}")
```
